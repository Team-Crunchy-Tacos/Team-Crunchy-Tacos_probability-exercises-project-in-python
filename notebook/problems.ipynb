## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np

box = {}

for i in range(60):
    if i < 10:
        box[i] = "White"
    elif (i > 9) and (i < 30):
        box[i] = "Red"
    else:
        box[i] = "Green"
print(box)
def draw(n_simulations = 1000):
    count_1 = 0
    count_2 = 0
    for i in range(n_simulations):
        color = []
        for i in range(5):
            color.append(box[np.random.randint(0, 59)])
        color = np.array(color)
        white_balls = sum(color == "White")
        red_balls = sum(color == "Red")
        green_balls = sum(color == "Green")
        if (white_balls == 3) and (red_balls == 2):
            count_1 += 1
        if (white_balls == 5) or (red_balls == 5) or (green_balls == 5):
            count_2 += 1
    return count_1 / n_simulations, count_2 / n_simulations
probabilities = draw(1000)
print(f"Probability for 3 white balls and 2 red balls: {np.round(probabilities[0] * 100, 2)}%")
print(f"Probability for all balls being the same color: {np.round(probabilities[1] * 100, 2)}%")





```
